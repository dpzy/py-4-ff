```python
xgdata[xgdata['Team'] == 'Watford'] ## Display 'Watford' from column 'Team'. n.b. this is a list, not tuple
```

```python
highg = xgdata[xgdata['G'] >= 15] ## Create Index where Column 'G' >=15
highg_or = highg.sort_values(by=['G', 'xG'], ascending=False) ## Sort-by G, then xG, in Descending Order

highg_or
```

```python
print(sorted(list(highg_or['Team']))) ## Print'Team's from highg_or list, sorted.
```
