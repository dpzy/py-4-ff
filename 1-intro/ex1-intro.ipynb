### 1.4 Adding Data to Dataframe <a name="1add"></a>

#### 1.4.1 Copying Dataframe 

`.copy` has a modifier named _deep_ with 2 modes.  
`deep = False` is a shallow copy where each data point is linked to parent dataframe.  
`deep = True` is a deep copy, ie. a totally new table.  

```python
xgdata_up = xgdata.copy(deep = False) # aka Shallow copy, values linked to precedent database
xgdata_up['GD'] = xgdata['G'] - xgdata['GA']
xgdata_up['xGD'] = xgdata['xG'] - xgdata['xGA']
xgdata_up['NPxGD'] = xgdata['NPxG'] - xgdata['xGA']

xgdata_up = xgdata_up.sort_values(by=['NPxGD'], ascending=False)
xgdata_up
```

#### 1.4.2 Indexing, Finding and Editing Data (`loc` and `iloc`)

`loc` gets the location of data by _label_ of the index. Labels are usually, but not always, character-string.  
- identifying column using `'columnnanme'` creates series format. `['columnname']` creates dataframe format.  

`iloc` gets the location of data by _absolute coordinates_ of the index. This is integer-only.  
    
> Alternatively, you may use...  
`dataframe.{columnname}`: generates a _series_ with index + column data.    
`dataframe['columnname']`: same as above.   
`datafreame[['columnname']]`: generates a _dataframe_ with index + column data.   

loc and iloc are multipurpose, hence recommended,

```python
#Start a Deep Copy
xgdata_deep = xgdata.copy(deep = True)
xgdata_deep
```

```python
#loc: 2d rectangular selection
xgdata_deep.iloc[:12, [0, 1, 2]]
```

```python
#loc: edit single value
xgdata_deep.loc[1, "G"] = 200
xgdata_deep
```

```python
#loc: edit multiple using rectanglular selection
xgdata_deep.loc[:19, ['xG', 'xGA']] = 200
xgdata_deep
```

```python
#loc.set_index: use another column as index

xgdata_deep.set_index('Team', inplace=True)
xgdata_deep
```

```python
#Using New Index
xgdata_deep.loc['Arsenal', 'G'] 
```

```python
#Performance-lite Alternative: .index and .pop
xgdata_deep2 = xgdata.copy(deep = True)
xgdata_deep2.index = xgdata_deep2['NPxG']  ##set Index
xgdata_deep2.pop('NPxG')  #pop: out old column
xgdata_deep2
```

## Plotting

### 2.1 Set the Dataframe Table <a name ="2plot"></a>

```python
xgplot = xgdata_up.copy(deep = True)
xgplot['GD-vs-xGD'] = xgplot['GD'] - xgplot['xGD']
xgplot.set_index('Team', inplace=True)
xgplot
```

### 2.2 Pyplot <a name="2pyplot"></a>

```python
# Import Pyplot (inside matplotlib)
from matplotlib import pyplot as plt #Impr
```

```python
#Arrange then Plot Graph
xgplot = xgplot.sort_values(by=['GD-vs-xGD'])  ##sort_values: to arrnage plot
plt.barh(xgplot.index, xgplot['GD-vs-xGD'])  ##barh: horizontal bar vs bar: vertical bar
plt.show()  #.show: important for pyplot
```

### 2.3 Seaborn Plot <a name="2seaborn"></a>

```python
#Library Dependencies: seaborn, pyplot, numpy(optional)
import seaborn as sns

sns.set_style('whitegrid')  
##.set_style includes: darkgrid, whitegrid, dark, white, ticks

sns.set_color_codes('colorblind')  
##.set_color_codes adjusts tint of default colors: pastel, dark, muted, deep, colorblind, bright
##.color_palette has more option https://seaborn.pydata.org/tutorial/color_palettes.html
```

```python
#Using Pyplots subplot function to scale
## Read more: https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplots
## Subplots Scaling: https://matplotlib.org/gallery/subplots_axes_and_figures/subplots_demo.html
f, ax = plt.subplots(figsize=(11, 10))

#Ploting Barplot for Categorical Variable y on Continuous Variable x.
sns.barplot(x='GD-vs-xGD', y=xgplot.index, data=xgplot, color='y')
ax.set(ylabel='', xlabel='Difference between GD and xGD')  ##ax from plt.subplots
sns.despine(left=True, bottom=True)  ##.despine: removes borders
```
